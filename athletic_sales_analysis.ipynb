### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# # Show the number products sold for region, state, and city.
# Group by 'region', 'state', and 'city' and sum the 'units_sold' column
grouped_sales = combined_df.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index()

# Rename the column 'units_sold' to 'Total_Products_Sold'
grouped_sales.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Sort the results in descending order to show the top 5 results
top_5_sales = grouped_sales.sort_values(by='Total_Products_Sold', ascending=False).head(5)

# Remove repeated values in 'region', 'state', and 'city' columns (blank out the repeated rows)
top_5_sales[['region', 'state', 'city']] = top_5_sales[['region', 'state', 'city']].apply(
    lambda col: col.where(col != col.shift(), '')
)

# Reset the index and display the top 5 results without index
top_5_sales = top_5_sales.reset_index(drop=True)
# Show the top 5 results.
top_5_sales.style.hide(axis='index')
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_sales = combined_df.pivot_table(
    index=['region', 'state', 'city'],
    values='units_sold',
    aggfunc='sum',
    fill_value=0
).reset_index()

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_sales.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Sort by 'Total_Products_Sold' in descending order
top_5_sales = pivot_sales.sort_values(by='Total_Products_Sold', ascending=False)

# Remove repeated values in 'region', 'state', and 'city'
top_5_sales[['region', 'state', 'city']] = top_5_sales[['region', 'state', 'city']].apply(
    lambda col: col.where(col != col.shift(), '')
)

# Reset the index to avoid displaying row numbers
top_5_sales = top_5_sales.reset_index(drop=True)

# Show the top 5 results.
top_5_sales.head().style.hide(axis='index')
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"

combined_df['total_sales'] = combined_df['units_sold'] * combined_df['price_per_unit']

# Group by 'region', 'state', and 'city' and sum the 'total_sales'
grouped_sales = (
    combined_df.groupby(['region', 'state', 'city'])['total_sales']
    .sum()
    .reset_index()
    .rename(columns={'total_sales': 'Total Sales'})
)

# # Sort the results by 'Total Sales' in descending order and get the top 5
top_5_sales = grouped_sales.sort_values(by='Total Sales', ascending=False).head(5)

# Leave subsequent rows blank if 'region', 'state', and 'city' values are repeated
for col in ['region', 'state', 'city']:
    top_5_sales[col] = top_5_sales[col].where(top_5_sales[col] != top_5_sales[col].shift(), '')

# Display the formatted top 5 results without row numbers
top_5_sales.style.hide(axis='index')
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
# Group by 'region', 'state', and 'city' and sum the 'total_sales' column
grouped_sales = combined_df.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index()

# Optional: Rename the column 'total_sales' to 'Total Sales'
grouped_sales.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Sort the results in descending order to show the top 5 results
top_5_sales = grouped_sales.sort_values(by='Total Sales', ascending=False).head(5)

# Leave subsequent rows blank if 'region', 'state', and 'city' values are repeated
for col in ['region', 'state', 'city']:
    top_5_sales[col] = top_5_sales[col].where(top_5_sales[col] != top_5_sales[col].shift(), '')
    
# Show the top 5 results.
top_5_sales.style.hide(axis='index')
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_footwear_sales = combined_df[combined_df["product"] == 'Women\'s Athletic Footwear']

# Display the filtered data to verify
womens_footwear_sales
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Group the filtered data by retailer, region, state, and city, and sum the 'units_sold' for each group
womens_footwear_sales_grouped = womens_footwear_sales.groupby(['retailer', 'region', 'state', 'city'])['units_sold'].sum().reset_index()

# Rename the 'units_sold' column to 'Womens_Footwear_Units_Sold'
womens_footwear_sales_grouped.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)

# Sort the data by 'Womens_Footwear_Units_Sold' in descending order to show the top 5 results
top_5_womens_footwear_sales = womens_footwear_sales_grouped.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head(5)

# Show the top 5 results.
top_5_womens_footwear_sales.style.hide(axis='index')
```

#### Using `pivot_table`

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
womens_footwear_sales = combined_df[combined_df['product'] == "Women's Athletic Footwear"]

# Create a pivot table with 'invoice_date' as the index and 'total_sales' as the values
sales_pivot= womens_footwear_sales.pivot_table(
    index='invoice_date', 
    values='total_sales', 
    aggfunc='sum', 
    fill_value=0
)

# Optional: Rename the 'total_sales' column to 'Total Sales'
sales_pivot.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Sort the pivot table by 'invoice_date' in ascending order to get the most recent dates last
sales_pivot_sorted = sales_pivot.sort_index(ascending=True)

# Show the last 10 days with the most sales
last_10_sales_days = sales_pivot_sorted.tail(10)

# Display the last 10 days with the most sales
last_10_sales_days

```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
sales_pivot.index = pd.to_datetime(sales_pivot.index, errors='coerce')

# Resample the pivot table into daily bins, summing the total sales for each day
sales_resampled = sales_pivot.resample('D').sum()

# Fill any missing days with 0 sales
sales_resampled = sales_resampled.fillna(0)

# Optional: Rename the 'total_sales' column to 'Total Sales'
sales_resampled.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Sort the resampled pivot table in descending order on 'Total Sales'
sales_resampled_sorted = sales_resampled.sort_values(by='Total Sales', ascending=False)

# Display the resampled and sorted pivot table
sales_resampled_sorted.head(10)

```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, summing the total sales for each week.
sales_resampled_weekly = sales_pivot.resample('W').sum()

# Optional: Rename the 'total_sales' column to 'Total Sales' if it's not done already
sales_resampled_weekly.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Sort the resampled pivot table in descending order on 'Total Sales'
sales_resampled_weekly_sorted = sales_resampled_weekly.sort_values(by='Total Sales', ascending=False)

# Display the resampled and sorted weekly pivot table
sales_resampled_weekly_sorted.head(10)
```
